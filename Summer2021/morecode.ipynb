```python
def frac_avail(df):
    listing = list(df.listing_id.unique())[0]
    counts = df.available.value_counts()
    #print(counts)
    try:
        frac = counts.iloc[0]/(counts.iloc[1] + counts[0])
        return frac
    except:
        if list(counts.index)[0] == 't':
            return 1
        else:
            return 'rented or lived in all the time'
    
    #return 
frac = usable_cal.groupby('listing_id').apply(frac_avail)
```

```python
fraction_rented = pd.merge(df['listings'], pd.DataFrame(frac), left_on = 'id', right_on = 'listing_id')
```
