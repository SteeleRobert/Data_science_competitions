```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('PlayerSpeedLabels.csv', index_col = [0])
```

```python
df.head()
```

```python
df.loc[df['Tournament'] == 'Dubai', 'Tournament'] = 1
df.loc[df['Tournament'] == 'Sydney', 'Tournament'] = 2
df.loc[df['Tournament'] == 'Commonwealth', 'Tournament'] = 3
df.loc[df['Tournament'] == 'Langford', 'Tournament'] = 4
df.loc[df['Tournament'] == 'Paris', 'Tournament'] = 5
df.loc[df['Tournament'] == 'World Cup', 'Tournament'] = 6

df.set_index('Tournament', inplace=True)
df = df.sort_values('Tournament')
```

```python
df.groupby('PlayerID')['Speed'].plot(figsize=(20,10))
plt.show()
```

```python
slopes = []
for i in (1,2,3,4,5,7,8,10,11,12,13,14,15,16):
    print(df[df['PlayerID'] == i]['Speed'].values)
    print(list(range(len(df[df['PlayerID'] == i]['Speed'].values))))
    slope, intercept = np.polyfit(list(range(len(df[df['PlayerID'] == i]['Speed'].values))), df[df['PlayerID'] == i]['Speed'].values, 1)
    slopes.append(slope)
```
